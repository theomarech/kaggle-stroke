# Présentation du jeux de données Stroke

Le jeux de données Stroke est un jeux de données structurés téléchargé sur [kaggle](https://www.kaggle.com/fedesoriano/stroke-prediction-dataset?select=healthcare-dataset-stroke-data.csv). L'origine de ce jeux de données est inconnue !  C'est un problème de classification binaire simple de présence absence d'AVC.
Il contient 11 variables explicatives et une variable dépendante qu'est la présence absence d'AVC.

Voici la présentation des variables : 
1) id: unique identifier
2) gender: "Male", "Female" or "Other"
3) age: age of the patient
4) hypertension: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension
5) heart_disease: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease
6) ever_married: "No" or "Yes"
7) work_type: "children", "Govt_jov", "Never_worked", "Private" or "Self-employed"
8) Residence_type: "Rural" or "Urban"
9) avg_glucose_level: average glucose level in blood
10) bmi: body mass index
11) smoking_status: "formerly smoked", "never smoked", "smokes" or "Unknown"*
12) stroke: 1 if the patient had a stroke or 0 if not
*Note: "Unknown" in smoking_status means that the information is unavailable for this patient

## Imputation des données BMI - Fumeur
On va tenter de créer un modèle pour prédire le BMI-Fumeur à partir des données et voir si on est meilleur qu'une imputation par la moyenne/fréquence.

### Préparation des données catégorielles : binarisation/OneHot/LabelEncoder

On transforme les données en format utilisable

```python
import numpy as np 

data.smoking_status = data.smoking_status.replace({'Unknown':np.nan,
                                                   'formerly smoked':1,
                                                   'never smoked':0,
                                                   'smokes':2})

data.gender = data.gender.replace({'Male':0,'Female':1})

data.ever_married = data.ever_married.replace({'Yes':1,'No':1})

data.Residence_type = data.Residence_type.replace({'Urban':0,'Rural':1})

df = pd.get_dummies(data,columns=["work_type"],prefix=["wkt"])

# suppression de la modalité other dans gender (1 seul représentant de la catégorie)
df = df[data.gender != 'Other']

df
```

### Visualisation des données quantitatives

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df['bmi'] = np.log2(df.bmi) # transformation de bmi en log pour l'approcher d'une loi normale

# transformation de bmi pour avoir une loi ~ normal 
g = sns.PairGrid(df, y_vars=["bmi", "avg_glucose_level","age"], x_vars=["stroke"], height=4)
g.map(sns.violinplot)

```

### Imputation des données BMI


```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV,cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score

df_bmi = df.loc[:,list(set(df.columns).difference({"stroke","smoking_status"}))].dropna()
X = df_bmi.loc[:,list(set(df_bmi.columns).difference({"bmi"}))]
y = df_bmi.loc[:,"bmi"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

params = [{'learning_rate':[0.5,0.1,0.01,0.001],
           'n_estimators':[10,100,500,1000,2000],
           'max_depth':[3,4,5,6,7],
           'max_features':[None,'auto', 'sqrt', 'log2']}]
          
gbr_bmi = GradientBoostingRegressor()

ggbr_gscv = RandomizedSearchCV(gbr_bmi,params,n_iter=30)
          
ggbr_gscv.fit(X_train,y_train)
```

```python
pred = ggbr_gscv.predict(X_test)
print(r2_score(y_test,pred))
# On obtient un R2 de 28% c'est mieux que rien !  
print(ggbr_gscv.best_params_)

bmi_model = GradientBoostingRegressor(n_estimators = ggbr_gscv.best_params_.get('n_estimators'),
                                     max_features = ggbr_gscv.best_params_.get('max_features'),
                                     max_depth = ggbr_gscv.best_params_.get('max_depth'),
                                     learning_rate = ggbr_gscv.best_params_.get('learning_rate'))
bmi_model.fit(X,y)

r2_score(y,bmi_model.predict(X)) #
```

On obtient un R2 de 35% environ sur l'échantillon de test ce qui est pas super mais pas si mal et 39% sur le modèle global. On va utiliser le modèle pour prédire les valeurs dans notre df.


```python
X_pred_bmi = df.loc[df.bmi.isna(),list(set(df.columns).difference({"stroke","smoking_status","bmi"}))]
X_pred_bmi["bmi"] = bmi_model.predict(X_pred_bmi)


df.loc[X_pred_bmi.index,'bmi'] = X_pred_bmi
```

### Imputation des données smoking status

```python
from sklearn.ensemble import GradientBoostingClassifier

df_smok = df.loc[:,list(set(df.columns).difference({"stroke"}))].dropna()
X = df_smok.loc[:,list(set(df_smok.columns).difference({"smoking_status"}))]
y = df_smok.loc[:,"smoking_status"].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

params = [{'learning_rate':[0.5,0.1,0.01,0.001],
           'n_estimators':[10,50,70,100,500,1000,2000],
           'max_depth':np.arange(3,9),
           'max_features':['auto','sqrt','log2']}]
          
```

```python
gbc_smok = GradientBoostingClassifier()

gbc_gscv_smok = RandomizedSearchCV(gbc_smok,params,n_iter=30)
          
gbc_gscv_smok.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score

y_prob = gbc_gscv_smok.predict_proba(X_test)

print(roc_auc_score(y_test, y_prob, multi_class="ovo",
                                  average="macro"))

print(gbc_gscv_smok.best_params_)

smoke_model = GradientBoostingClassifier(n_estimators = gbc_gscv_smok.best_params_.get('n_estimators'),
                                         max_features = gbc_gscv_smok.best_params_.get('max_features'),
                                         max_depth=gbc_gscv_smok.best_params_.get('max_depth'),
                                         learning_rate=gbc_gscv_smok.best_params_.get('learning_rate')).fit(X,y)

y_prob = gbc_gscv_smok.predict_proba(X)

roc_auc_score(y, y_prob, multi_class="ovo",
                                  average="macro")

X_pred_smoke = df.loc[df.smoking_status.isna(),list(set(df.columns).difference({"stroke","smoking_status"}))]
X_pred_smoke["smoking_status"] = smoke_model.predict(X_pred_smoke)


df.loc[X_pred_smoke.index,'smoking_status'] = X_pred_smoke
```

```python
from sklearn.metrics import precision_score, f1_score

permut = 10
n_iter = 50
params = [{'learning_rate':[0.5,0.1,0.01,0.005,0.001],
               'n_estimators':[10,30,50,70,100,500,1000,2000],
               'max_depth':[3,4,5,6,7],
               'subsample':[0.1,0.3,0.6,0.8,1.0],
               'min_weight_fraction_leaf':[0.01,0.03],
               'max_features':[None,'auto','sqrt','log2'],
               'min_weight_fraction_leaf':[0.001,0.002,0.003,0.004,0.006,0.008,0.01,0.02,0.03,0.04]}]
    
def permutation_score(df, permut = 3, n_iter = 1):
    best_param_ = []
    auc_score_ = []
    f1_score_ = []
    precision = []
    for _ in range(permut):
        x_minority = df.loc[df.stroke == 1,:] # les données de stroke +
        x_majority = df.loc[df.stroke == 0,:] # les données de stroke -

        x_majority_permut = x_majority.sample(frac=x_minority.shape[0]/df.shape[0])
        data_permut = pd.concat([x_majority_permut,x_minority])
        X = data_permut.loc[:,data_permut.columns!='stroke']
        y = data_permut.loc[:,'stroke']

        X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, stratify = y)
        gbr = GradientBoostingClassifier()
        gbr_stroke = RandomizedSearchCV(gbr,params,n_iter = n_iter).fit(X_train,y_train)

        pred_proba = gbr_stroke.predict_proba(X_test)[:,1]
        pred = gbr_stroke.predict(X_test)

        best_param_.append(gbr_stroke.best_params_)
        auc_score_.append(roc_auc_score(y_test,pred_proba))
        f1_score_.append(f1_score(y_test,pred))
        precision.append(precision_score(y_test,pred))
        
    return best_param_,auc_score_,f1_score_, precision

X = df.loc[:,df.columns!='stroke']
y = df.loc[:,'stroke']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,stratify=y)

data_train = pd.concat([X_train,y_train],axis=1)

param, auc, f1, prec = permutation_score(data_train, permut = permut, n_iter = n_iter)
```

```python


gbr = GradientBoostingClassifier(subsample = param.get('subsample'), 
                                 n_estimators = param.get('n_estimators'),
                                 min_weight_fraction_leaf = param.get('min_weight_fraction_leaf'),
                                 max_features = param.get('max_features'),
                                 max_depth = param.get('max_depth'),
                                 learning_rate = param.get('learning_rate'))

x_minority = data_train.loc[data_train.stroke == 1,:] # les données de stroke +
x_majority = data_train.loc[data_train.stroke == 0,:] # les données de stroke -

x_majority_permut = x_majority.sample(frac=x_minority.shape[0]/data_train.shape[0])
data_permut = pd.concat([x_majority_permut,x_minority])
X = data_permut.loc[:,data_permut.columns!='stroke']
y = data_permut.loc[:,'stroke']

gbr.fit(X,y)
```
